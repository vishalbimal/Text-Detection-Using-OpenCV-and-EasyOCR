```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

# 1. read in images or video

```python
image_path="C:\\Users\\visha\\OneDrive\\Desktop\\image.jpg"
```

```python
reader=easyocr.Reader(['en'],gpu=False)
result=reader.readtext(image_path)
result
```

# draw results

```python
top_left=tuple(result[0][0][0])
bottom_right=tuple(result[0][0][2])
text=result[0][1]
font=cv2.FONT_HERSHEY_SIMPLEX
```

```python
img=cv2.imread(image_path)
img=cv2.rectangle(img,top_left,bottom_right,(0,255,0),5)
img=cv2.putText(img,text,top_left,font,0.5,(255,255,255),2,cv2.LINE_AA)
plt.imshow(img)
plt.show()
```

```python
image_path="C:\\Users\\visha\\OneDrive\\Desktop\\london-uk-2nd-december-2021-north-middlesex-university-hospital-signage-with-directions-to-different-hospital-areas-2HEB13M.jpg"
```

```python
reader=easyocr.Reader(['en'],gpu=False)
result=reader.readtext(image_path)
result
```

```python
img=cv2.imread(image_path)
for detection in result:
    top_left=tuple([int(val) for val in detection[0][0]])
    bottom_right=tuple([int(val) for val in detection[0][2]])
    text=detection[1]
    font=cv2.FONT_HERSHEY_SIMPLEX
    img=cv2.rectangle(img,top_left,bottom_right,(0,255,0),5)
    img=cv2.putText(img,text,top_left,font,0.5,(255,255,255),2,cv2.LINE_AA)
plt.figure(figsize=(10,15))
plt.imshow(img)
plt.show()
```
